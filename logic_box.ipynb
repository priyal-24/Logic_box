```python
# Pattern Generator and Number Analyzer

print("=" * 60)
print("      WELCOME TO PATTERN GENERATOR & NUMBER ANALYZER")
print("=" * 60)
print("Generate Patterns | Analyze Numbers | Learn Logic")
print("-" * 60)

while True:

    print("\nMAIN MENU")
    print("-" * 20)
    print("1. Generate a Pattern")
    print("2. Analyze a Range of Numbers")
    print("3. Exit")

    choice = input("Enter your choice (1-3): ")

    if choice == "1":

        rows = int(input("Enter the number of rows: "))

        if rows <= 0:
            print("Invalid row count! Please enter a positive number.")
            continue

        print("\n--- Right-Angled Triangle Pattern ---")

        for i in range(1, rows + 1):
            for j in range(i):
                print("*", end="")
            print()

    elif choice == "2":

        start = int(input("Enter the start of the range: "))
        end = int(input("Enter the end of the range: "))

        if start > end:
            print("Invalid range! Start should be less than or equal to End.")
            continue

        total = 0

        print("\n--- Number Analysis Report ---")

        for num in range(start, end + 1):

            # pass statement
            if start == end:
                pass

            if num % 2 == 0:
                print(f"{num} is Even")
            else:
                print(f"{num} is Odd")

            total += num

        print(f"\nSum of all numbers from {start} to {end} = {total}")

    elif choice == "3":

        print("\n" + "=" * 60)
        print("Thank you for using Pattern Generator & Number Analyzer.")
        print("Program terminated successfully.")
        print("Have a great day!")
        print("=" * 60)

        break

    else:
        print("Invalid choice! Please try again.")
```
